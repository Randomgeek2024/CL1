```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
data = pd.read_csv("data.csv", encoding='latin1')
print("âœ… Dataset Loaded")
print("Shape:", data.shape)
print("Index:", data.index, "\n")

# Step 3: Inspect dataset
print("Missing values per column:\n", data.isnull().sum(), "\n")
data.info()
print("Last 5 rows:\n", data.tail(), "\n")

# Step 4: Set seaborn style and figure size
sns.set(style="ticks", rc={'figure.figsize': (20, 15)})

# Step 5: Mean NO2 values by state
state_no2_means = data.groupby('state')['no2'].mean()
plt.figure(figsize=(12, 6))
plt.bar(state_no2_means.index, state_no2_means.values, color='blue')
plt.xlabel('State')
plt.ylabel('Mean NO2 Value')
plt.title('Mean NO2 Values by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Step 6: Mean SO2 values by state
state_so2_means = data.groupby('state')['so2'].mean()
plt.figure(figsize=(12, 6))
plt.bar(state_so2_means.index, state_so2_means.values, color='green')
plt.xlabel('State')
plt.ylabel('Mean SO2 Value')
plt.title('Mean SO2 Values by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
